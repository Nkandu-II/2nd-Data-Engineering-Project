## Size column

```python
print(play_store.Size.sort_values().unique())
```

```python
(play_store['Size'] == "Varies with device").sum()
```

```python
def convert(size):
    if size == "Varies with device":
        return np.nan
    elif size[-1] == 'M':
        return float(size[:-1]) * 1024
    else:
        return float(size[:-1])

play_store['Size'] = play_store['Size'].apply(convert)
```

# Second Week

1- Display (table) the unique names of all categories 
    
2- Plot a bar chart for categories with the total number of installing numbers in each category 
    
3- Plot a bar chart for the total prices of each paid app in each category ( the sum of all prices in the same category)
    
4- Plot a bar chart of the total profit of each category by multiplying the price by the number of installs

```python
# 1- Display (table) the unique names of all categories

unique_categories = play_store['Category'].unique()
playstore_categories = pd.DataFrame({'Category': unique_categories})
print(playstore_categories)
```

```python
# Plot a bar chart for categories with the total number of installing numbers in each category
# grouped = play_store.groupby('Category').sum()
# grouped.plot(kind='bar', y='Installs')

grouped = play_store.groupby('Category').sum()
categories = grouped.index
installs = grouped['Installs']

plt.bar(categories, installs)
plt.xlabel('Category')
plt.ylabel('Total Installs')
plt.title('Total number of Installations by Category')

plt.xticks(rotation=90)
plt.yscale('log')

plt.show()
```

```python
# 3- Plot a bar chart for the total prices of each paid app in each category (the sum of all prices in the same category)
grouped = play_store.groupby('Category').sum()
categories = grouped.index
prices = grouped['Price']

plt.bar(categories, prices)
plt.xlabel('Category')
plt.ylabel('Total Price')
plt.title('Total Price by Category')

plt.yscale('log')
plt.xticks(rotation=90)

plt.show()
```

```python
# 4- Plot a bar chart of the total profit of each category by multiplying the price by the number of installs
play_store['Profit'] = play_store['Price'] * play_store['Installs']
play_store.groupby('Category')['Profit'].sum().plot.bar()

plt.xlabel('Category')
plt.ylabel('Total Profit')
plt.title('Total Profit by Category')

plt.yscale('log')

plt.show()
```

```python
#  Display a table and a chart for top 5 download apps in each category free and paid (if there is paid)

# Filter only the relevant columns
play_store_t5 = play_store[['Category', 'App', 'Type', 'Installs']]

# Sort the data by category and number of installs
play_store_t5 = play_store_t5.sort_values(by=['Category', 'Installs'], ascending=[True, False])

# Group the data by category and type
grouped = play_store_t5.groupby(['Category', 'Type'])

# Get the top 5 apps in each category and type
top_apps = grouped.head(5)

# Create a bar chart to display the top 5 downloaded apps in each category and type
fig = px.bar(top_apps, x='App', y='Installs', color='Type')


# Show the chart
fig.show()
```

```python
#  Display a sunburst plot grouping by categories
# Group the data by categories
grouped = play_store.groupby(['Category', 'App']).sum().reset_index()

# Create the sunburst plot
fig = px.sunburst(grouped, path=['Category', 'App'], values='Installs',
                  color='Category', hover_data=['Installs'],
                  color_continuous_scale='Viridis')

# Show the plot
fig.show()
```

```python
#- Apply sliding window technique  and store the patches in a 2D list

def sliding_window(data, window_size):
    patches = []
    for i in range(len(data) - window_size + 1):
        patch = data[i:i + window_size]
        patches.append(patch)
    return patches

window_size = 100
patches = sliding_window(play_store, window_size)
print(len(patches))
```
